```swift
// Custom differentiable type.
struct Model: Differentiable {
    var w: Float
    var b: Float
    func applied(to input: Float) -> Float {
        return w * input + b
    }
}

// Differentiate using `gradient(at:_:in:)`.
let model = Model(w: 4, b: 3)
let input: Float = 2
let (ğ›model, ğ›input) = gradient(at: model, input) { model, input in
    model.applied(to: input)
}

print(ğ›model) // Model.TangentVector(w: 2.0, b: 1.0)
print(ğ›input) // 4.0
```

## How it works

Swift's `Differentiable` protocol enables automatic differentiation:

1. **`Differentiable` conformance**: Types that conform to `Differentiable` can be differentiated
2. **`gradient(at:in:)`**: Computes gradients of a function with respect to its inputs
3. **`TangentVector`**: Auto-synthesized type that holds the gradients

The gradient of `w * input + b` with respect to:
- `w` is `input` (which is 2.0)
- `b` is 1.0 (constant term)
- `input` is `w` (which is 4.0)
